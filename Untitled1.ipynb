```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
columns=['Sepal length','Sepal width','Petal length','Petal width','Class_labels']
df=pd.read_csv("IRIS.csv",names=columns)
df.head()
```

```python
df.describe()
df.head()
```

```python
for col in ['Sepal length','Sepal width','Petal length','Petal width']:
  df[col]=pd.to_numeric(df[col],errors='coerce')
df['Sepal length']=df['Sepal length'].fillna(df['Sepal length'].median())
df['Sepal width']=df['Sepal width'].fillna(df['Sepal width'].median())
df['Petal length']=df['Petal length'].fillna(df['Petal length'].median())
df['Petal width']=df['Petal width'].fillna(df['Petal width'].median())

#visualizing
#sns.pairplot(df,hue='Class_labels')
```

```python
#Seprate feature and target
data=df.values
X=data[:,0:4]
Y=data[:,4]
print(X)
```

```python
#split the data into train and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction1=model_svc.predict(X_test)
#calculate accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1 )*100)
for i in range(len(prediction1)):
  print(Y_test[i],prediction1[i])
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model_Lr=LogisticRegression()
model_Lr.fit(X_train,Y_train)
```

```python
prediction2=model_Lr.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1))
for i in range(len(prediction2)):
  print(Y_test[i],prediction2[i])
```

```python
#decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
model_dtc=DecisionTreeClassifier()
model_dtc.fit(X_train,Y_train)
```

```python
prediction3=model_dtc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction3))
for i in range(len(prediction3)):
  print(Y_test[i],prediction3[i])
```

```python
#classicification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction3))
```
